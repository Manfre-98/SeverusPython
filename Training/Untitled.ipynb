```python
# univariate cnn example
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
import pandas as pd
import numpy as np
import keras
```

```python
c = pd.read_csv("resNew.csv")

arr = np.empty((12395,81),dtype='float32')

print(arr.shape)

i=0
n=0

for riga in c["state"]:
    n=0    
    for el in riga:
        arr[i,n]=el        
        n+=1
    i+=1    
    
arr= arr/3
```

```python
assert not np.any(np.isnan(arr))
```

```python
out = np.empty((12395,2),dtype=object)

i=0

for i in range(12395):
    out[i,0]=abs(c['w'][i] - c['b'][i])
    if((c['w'][i] - c['b'][i])>0):
        out[i,1]=0
    else:
        out[i,1]=1

out=out.astype('float32')


print(out.shape)
```

```python
assert not np.any(np.isnan(out))
```

```python
# choose a number of time steps
n_steps = 3
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = arr.reshape((arr.shape[0], arr.shape[1], n_features))
# define model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(81, n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(2))
opt = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=opt, loss='mse')
# fit model
mcp_save = keras.callbacks.ModelCheckpoint('net.h5', save_best_only=True, monitor='val_loss', mode='min')
model.fit(X, out, epochs=1000, verbose=1, validation_split = 0.2, callbacks=[mcp_save])
model.save('net.h5')

```
